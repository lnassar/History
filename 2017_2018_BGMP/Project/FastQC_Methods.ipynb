# Quality Analysis of Data

## FastQC

### Gularis

First the files were unzipped into our shared BGMP directory:
```
jar -xf /home/lnassar/Project/RNA_Seq_data/gularis_RNASeq.zip
jar -xf /home/lnassar/Project/RNA_Seq_data/inornata_RNASeq.zip
```
    Note: unzip did not work in this instance as the files were too large
    
The gularis directory had two separate runs, with each of them containing 28 files. The 2086 was subjected to a ribosomal depletion an the 2088 was a poly A selection. Also worth noting is the inclusion of a samples.txt file with information on the run/samples including tissue of origin. Of those 28 files half were R1, half R2. Additionally they were broken down into what seems the first half of a read, and the second. So the S_N_N nomenclature means:

S_**1**_1_Index = The bold '1' is the first part of the read

S_1_**1**_Index = The bold '1' here signifies it is the forward strand

So 28/4 = There are 7 unique stranded reads in each of the two runs.

A slurm script was made using FastQC on each individual file and outputing it to a unique folder. Following are the non-header parts of the scripts
```
#For Gularis 2036 (this is one of the two sequence identifiers for the runs, the other is 2038)
ml fastqc
fastqc -o /projects/bgmp/lizards/Gularis_RNAseq/gularis_RNASeq/MOLNG2036/FastQC_Gularis_2036 *.fastq.gz
```
```
#For Gularis 2038
ml fastqc
fastqc -o /projects/bgmp/lizards/Gularis_RNAseq/gularis_RNASeq/MOLNG2038/FastQC_Gularis_2038 *.fastq.gz
```
    Note: Version 0.11.5 of FastQC was used for this analyses

### Inornata

Similar to gularis first the files were unzipped into the afformentioned directory. In this directory all of it was done in a single run, so you have half the data as the gularis data set. The data however is broken into 32 files instead of 28, so instead of 7 unique stranded reads for inornata we have 8 unique stranded reads. The titles are also easier to read with R1/R2 denotation and origin of the tissue.

Like Gularis, a slurm script was submitted. The same FastQC version as previous was used. The non-header was as follows:
```
ml fastqc
fastqc -o /projects/bgmp/lizards/Inornata_RNAseq/inornata_RNASeq/FastQC_Inornata *.fastq.gz
```

## MultiQC

MultiQC is a software that aggregates results from bioinformatics analyses across many samples into a single report. In our case it is being used to combine all of the FastQC reports from each respective sequencing run. First it was installed with the following commands:

```
git clone https://github.com/ewels/MultiQC.git
cd MultiQC
python setup.py install
```

After it was installed all the FastQC files were copied to the local directory and from within each directory MultiQC was run with the simple command:

```
multiqc .
```
    Note: Version used was v1.3.dev0
