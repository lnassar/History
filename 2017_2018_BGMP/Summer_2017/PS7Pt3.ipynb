```python
#Open my files and set variables
my_file = open("zebrafish-v-humandb.txt")
my2_file = open("human-v-zebrafishdb.txt")
best_proteinH = {}
best_proteinZ = {}

#Iterate through every line in the  file, and keep only the queryID with the lowest evalue, and add that to our dictionary
def MakeUniqDicH (blastfile):
    n=-1
    k=0
    best_protein = {}
    for each in blastfile:
        n+=1
        each = each.split('\t')
        if n == 0:
            QueryProteinID = each[0][0:18]
            SubjectID = each[1][0:15]
            evalue = float(each[10])
        elif QueryProteinID == each[0][0:18]:
            if evalue < float(each[10]):
                best_protein[QueryProteinID]=SubjectID
                k=k+1
            elif evalue == float(each[10]):
                k=k+1
            elif evalue > float(each[10]):
                print("Houston: We have a problem: A lower GeneID evalue shoudl not be less than a previous value")
        elif QueryProteinID != each[0][0:18]:
            if k == 0:
                if QueryProteinID in best_protein:
                    print('Houston, we have a problem: Your dictionary is adding values that are already in there')
                elif QueryProteinID not in best_protein:
                    best_proteinH[QueryProteinID]=SubjectID
            QueryProteinID = each[0][0:18]
            SubjectID = each[1][0:15]
            evalue = float(each[10])
            k=0
    return best_protein

def MakeUniqDicZ (blastfile):
    n=-1
    k=0
    best_protein = {}
    for each in blastfile:
        n+=1
        each = each.split('\t')
        if n == 0:
            QueryProteinID = each[0][0:15]
            SubjectID = each[1][0:18]
            evalue = float(each[10])
        elif QueryProteinID == each[0][0:15]:
            if evalue < float(each[10]):
                best_protein[QueryProteinID]=SubjectID
                k=k+1
            elif evalue == float(each[10]):
                k=k+1
            elif evalue > float(each[10]):
                print("Houston: We have a problem: A lower GeneID evalue shoudl not be less than a previous value")
        elif QueryProteinID != each[0][0:15]:
            if k == 0:
                #best_protein[QueryProteinID]=SubjectID
                if QueryProteinID in best_protein:
                    print('Houston, we have a problem: Your dictionary is adding values that are already in there')
                elif QueryProteinID not in best_protein:
                    best_proteinH[QueryProteinID]=SubjectID
            QueryProteinID = each[0][0:15]
            SubjectID = each[1][0:18]
            evalue = float(each[10])
            k=0
    return best_protein

#Run the function of both files and same them into dictionary variables
best_proteinZ = MakeUniqDicH(my_file)
best_proteinH = MakeUniqDicZ(my2_file)

#Print length of dictionaries, and find RBH, then display the amount of hits. Then create a new RBH dictionary.
RBH_Dic = {} #Format: Protein ID for human, first entry: zebra protein ID
for keys,values in best_proteinZ.items():
    if values in best_proteinH and best_proteinH[values] == keys:
        RBH_Dic[values]=keys

#Print statements of stats and quality assurance
print("The length of the unique E-Value Human to Zebrafish dictionary is",len(best_proteinH))
print("The length of the unique E-Value Zebrafish to Human dictionary is",len(best_proteinZ))
print("The total number of RBH hits is",len(RBH_Dic))

my_file.close()
my2_file.close()
```
