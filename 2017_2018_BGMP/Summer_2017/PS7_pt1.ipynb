```python
import re

#Set my file variables and open by dictionary
#tablefile=open("zebrafish_table.txt.txt")
tablefile=open("human_table.txt")
#proteinfile=open("Danio_rerio.GRCz10.pep.all.fa")
proteinfile=open("Homo_sapiens.GRCh38.pep.all.fa")
tableIDs = {}

#Test file for QA
#proteinfile=open("test.txt")

#Populate my dictionary keys with the geneIDs that also match for having a proteinID
for lines in tablefile:
    lines=lines.split("\t")
    if len(lines[2]) != 1 and len(lines[0]) != 14:
        tableIDs[lines[0].strip("\n")] = []

#Iterate through the file and make it into a list
fasta = []
for lines in proteinfile:
   lines = lines.strip("\n")
   fasta.append(lines)

#Iterate through the list of the file and match gene IDs to the dictionary and find
#and save the sequences, replacing them and keeping only the longest
n=-1
for each in fasta:
    n+=1
#These first 2 conditions are to account for the first line
####THE GENE AND PROTEIN IDS NEED TO BE CHANGED BETWEEN FISH AND HUMANS
    if n==0:
        header=each
        #proteinID=header[1:19] #TURN THIS ON FOR FISH PROTEOME
        proteinID=header[1:16] #TURN THIS ON FOR HUMAN PROTEOME
        each=each.split(" ")
        #geneID = each[3][5:23] #TURN THIS ON FOR FISH PROTEOME
        geneID = each[3][5:20] #TURN THIS ON FOR HUMAN PROTEOME
    elif n==1:
        seq=each
    elif re.search("^>",each):
        if geneID in tableIDs.keys() and tableIDs[geneID]==[]:
            tableIDs[geneID].append(header)
            tableIDs[geneID].append(seq)
        elif len(seq) > len(tableIDs[geneID][1]):
            tableIDs[geneID][0] = header
            tableIDs[geneID][1] = seq
        header = each
        #proteinID=header[1:19] #TURN THIS ON FOR FISH PROTEOME
        proteinID=header[1:16] #TURN THIS ON FOR HUMAN PROTEOME
        each=each.split(" ")
        #geneID = each[3][5:23] #TURN THIS ON FOR FISH PROTEOME
        geneID = each[3][5:20] #TURN THIS ON FOR HUMAN PROTEOME
        seq = ""
#Following loop is to account for the last line in the file
    elif len(fasta) == n+1:
        seq = seq + each
        if geneID in tableIDs.keys() and tableIDs[geneID]==[]:
            tableIDs[geneID].append(header)
            tableIDs[geneID].append(seq)
        elif len(seq) > len(tableIDs[geneID][1]):
            tableIDs[geneID][0] = header
            tableIDs[geneID][1] = seq

#Add the sequence lines
    else:
        seq = seq + each

#Output our fasta format
for keys, values in tableIDs.items():
   print(tableIDs[keys][0]+"\n"+tableIDs[keys][1])

#QA and file closing
#print("the number of lines in the file is",n)
#print("The length of your fasta is",len(fasta))
#print(tableIDs)
tablefile.close()
proteinfile.close()
```
