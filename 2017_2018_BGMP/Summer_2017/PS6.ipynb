```python
##PS6 Part 1
my_file = open("contigs.fa")
#Specify your used kmer
kmer = 31
contigs = []
kmer_length = []
kmer_coverage = []
total_contigs = 0
for each in my_file:
    if each.startswith(">"):
        total_contigs = total_contigs + 1
        contigs.append(each.strip('\n'))

#Isolate the kmer counts and depth coverage and convert to sequence length
for each in contigs:
    current=each.split("_")
    ###Append the converted kmer length with a kmer of 49, Kcount=L-K+1, in the case of #1: 3317=L-49+1
    kmer_length.append(int(current[3])+(kmer-1))
    kmer_coverage.append((float(current[5])*(float(current[3])+(kmer-1))/((float(current[3])+(kmer-1))-(kmer-1))))
    
#Caculate the total length of genome accross all contigs
total_genome = sum(kmer_length)
print("The total length  of the genome accross all contigs is",total_genome)

#Find the mean contig
mean_contig=sum(kmer_length) / len(kmer_length)
print("The mean contig value is",mean_contig)

#sort the kmer length list and print the largest value
kmer_length = sorted(kmer_length)
print("The largest contig length is",kmer_length[total_contigs-1])

#Calculate the mean depth of coverage
mean_coverage=sum(kmer_coverage) / len(kmer_coverage)
print("The mean coverage value is",mean_coverage)

#Calculate the N50 value, or median value
print("The N50 value for this assembly is",kmer_length[int(total_contigs/2)])

#print(contigs)
print("There total number of contigs present in the file is",total_contigs)

#Bin the kmer lengths using floor division and a dictionary
binval = {}
for each in kmer_length:
    if each//100 not in binval:
        binval[each//100]=1
    else:
        binval[each//100]=binval[each//100]+1

#Print my binned contigs
print ("# Contig length \t Number of contigs in category")
for keys,values in binval.items():
    print(str(int(keys)*100)+"\t"+str(values))
my_file.close()
```
